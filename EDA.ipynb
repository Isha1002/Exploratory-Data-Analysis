```python
#missing values
titanic_df.isna().sum()
```

```python
median_age=titanic_df['Age'].median()
```

```python
titanic_df['Age']=titanic_df['Age'].fillna(median_age)
```

```python
titanic_df['Age']
```

```python
titanic_df.isna().sum()
```

```python
mode=titanic_df['Embarked'].mode()
mode
```

```python
titanic_df['Embarked']=titanic_df['Embarked'].fillna(mode[0])
```

```python
titanic_df.isna().sum()
```

```python
titanic_df['Has_cabin']=titanic_df['Cabin'].notna().astype(int)
titanic_df.drop('Cabin',axis=1,inplace=True)
titanic_df.head()
```

```python
titanic_df['Has_cabin'].value_counts()
```

```python
#Univariate Analysis
print("Categorical features")
fig,axes=plt.subplots(2,4,figsize=(16,12))
fig.suptitle('Categorical features')

sns.countplot(ax=axes[0][0],data=titanic_df,x='Sex').set_title("Gender Distribution")
sns.countplot(ax=axes[0][1],data=titanic_df,x='Pclass').set_title("Pclass Distribution")
sns.countplot(ax=axes[0][2],data=titanic_df,x='Embarked').set_title("Embarked Distribution")
sns.countplot(ax=axes[1][0],data=titanic_df,x='Survived').set_title("Survived Distribution")
sns.countplot(ax=axes[1][1],data=titanic_df,x='Has_cabin').set_title("Has_cabin Distribution")
plt.tight_layout()
plt.show()
```

```python
#Numberical features
print("Numerical features")
Fig,axes = plt.subplots(1,2,figsize=(16,12))
fig.suptitle('Numerical features')

sns.histplot(ax=axes[0],data=titanic_df,x='Age',kde=True,bins=30).set_title("Age Distribution")
sns.histplot(ax=axes[1],data=titanic_df,x='Fare',kde=True,bins=40).set_title("Fare Distribution")
plt.show()
```

"Bivariate analysis

```python
fig,axes =plt.subplots(2,2,figsize=(12,10))
fig.suptitle('Bivariate Analysis')
#pclass vs survived:
sns.barplot(ax=axes[0,0],x='Pclass',y='Survived',data=titanic_df).set_title("Pclass vs Survived")
#sex vs survived:
sns.barplot(ax=axes[0,1],x='Sex',y='Survived',data=titanic_df).set_title("Sex vs Survived")
#embarked vs survived
sns.barplot(ax=axes[1,0],x='Embarked',y='Survived',data=titanic_df).set_title("Embarked vs Survived")
#cabin vs survived
sns.barplot(ax=axes[1,1],x='Has_cabin',y='Survived',data=titanic_df).set_title("Has_cabin vs Survived")
plt.tight_layout()
plt.show()
```
