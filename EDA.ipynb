```python
#missing values
titanic_df.isna().sum()
```

```python
median_age=titanic_df['Age'].median()
```

```python
titanic_df['Age']=titanic_df['Age'].fillna(median_age)
```

```python
titanic_df['Age']
```

```python
titanic_df.isna().sum()
```

```python
mode=titanic_df['Embarked'].mode()
mode
```

```python
titanic_df['Embarked']=titanic_df['Embarked'].fillna(mode[0])
```

```python
titanic_df.isna().sum()
```

```python
titanic_df['Has_cabin']=titanic_df['Cabin'].notna().astype(int)
titanic_df.drop('Cabin',axis=1,inplace=True)
titanic_df.head()
```

```python
titanic_df['Has_cabin'].value_counts()
```

```python
#Univariate Analysis
print("Categorical features")
fig,axes=plt.subplots(2,4,figsize=(16,12))
fig.suptitle('Categorical features')

sns.countplot(ax=axes[0][0],data=titanic_df,x='Sex').set_title("Gender Distribution")
sns.countplot(ax=axes[0][1],data=titanic_df,x='Pclass').set_title("Pclass Distribution")
sns.countplot(ax=axes[0][2],data=titanic_df,x='Embarked').set_title("Embarked Distribution")
sns.countplot(ax=axes[1][0],data=titanic_df,x='Survived').set_title("Survived Distribution")
sns.countplot(ax=axes[1][1],data=titanic_df,x='Has_cabin').set_title("Has_cabin Distribution")
plt.tight_layout()
plt.show()
```

```python
#Numberical features
print("Numerical features")
Fig,axes = plt.subplots(1,2,figsize=(16,12))
fig.suptitle('Numerical features')

sns.histplot(ax=axes[0],data=titanic_df,x='Age',kde=True,bins=30).set_title("Age Distribution")
sns.histplot(ax=axes[1],data=titanic_df,x='Fare',kde=True,bins=40).set_title("Fare Distribution")
plt.show()
```

"Bivariate analysis

```python
fig,axes =plt.subplots(2,2,figsize=(12,10))
fig.suptitle('Bivariate Analysis')
#pclass vs survived:
sns.barplot(ax=axes[0,0],x='Pclass',y='Survived',data=titanic_df).set_title("Pclass vs Survived")
#sex vs survived:
sns.barplot(ax=axes[0,1],x='Sex',y='Survived',data=titanic_df).set_title("Sex vs Survived")
#embarked vs survived
sns.barplot(ax=axes[1,0],x='Embarked',y='Survived',data=titanic_df).set_title("Embarked vs Survived")
#cabin vs survived
sns.barplot(ax=axes[1,1],x='Has_cabin',y='Survived',data=titanic_df).set_title("Has_cabin vs Survived")
plt.tight_layout()
plt.show()
```

```python
#Feature engineering
```

```python
#creating a family size feature
titanic_df['Family_size']=titanic_df['SibSp']+titanic_df['Parch']+1
#creating is alone feature
titanic_df['Is_alone']=0
titanic_df.loc[titanic_df['Family_size']==1,'Is_alone']=1

print("Created family size and isalone features")
titanic_df[['Family_size','Is_alone']].head()
```

```python
#Analyzing the family related features against survival
fig,axes =plt.subplots(1,2,figsize=(12,10))
sns.barplot(ax=axes[0],x='Family_size',y='Survived',data=titanic_df).set_title("Family_size vs Survived")
sns.barplot(ax=axes[1],x='Is_alone',y='Survived',data=titanic_df).set_title("Is_alone vs Survived")
plt.tight_layout()
plt.show()
```

```python
#Extracting title from the name column
titanic_df['Title']=titanic_df['Name'].str.extract('([A-Za-z]+)\.',expand=True)
titanic_df['Title'].value_counts()
```

```python
titanic_df['Title']=titanic_df['Title'].replace(['Rev','Col','Mile','Major','Don','Dr','Lady','Sir','Capt','Countess','Jonkheer'],'Rare')
titanic_df['Title']=titanic_df['Title'].replace(['Ms','Mlle'],'Miss')
titanic_df['Title']=titanic_df['Title'].replace(['Mme'],'Mrs')
titanic_df['Title'].value_counts()
plt.figure(figsize=(6,6))
sns.barplot(x='Title',y='Survived',data=titanic_df)
plt.show()
```

```python
#Multivariate analysis
sns.catplot(x='Title',y='Survived',hue='Sex',kind='bar',data=titanic_df)
plt.title('Title vs Survived (hue=Sex)')
plt.show()
```

```python
#violing plot to see age distribution by sex and survival status
plt.figure(figsize=(10,6))
sns.violinplot(x='Sex',y='Age',hue='Survived',split=True,data=titanic_df)
plt.title('Age Distribution by Sex and Survival')
plt.show()
```

```python
#Correlation analysis
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(titanic_df.drop(['Name', 'Ticket'], axis=1).corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
